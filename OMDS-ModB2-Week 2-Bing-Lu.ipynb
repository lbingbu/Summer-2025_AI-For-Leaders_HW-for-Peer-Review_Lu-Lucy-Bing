```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = "/Users/lubing/Desktop/marketing_campaign_13Apr_full dataset.xlsx"
df = pd.read_excel(file_path)

df.info()
df.describe(include='all')
```

```python
# Are there any duplicate rows?
duplicate_rows = df.duplicated().sum()
print(f"Duplicate rows: {duplicate_rows}")
```

```python
# Are there any duplicate values in a given column?
# Check for duplicate values within each object (categorical) column
for col in df.select_dtypes(include='object').columns:
    duplicates = df[col].duplicated().sum()
    print(f"Duplicate values in column '{col}': {duplicates}")
```

```python
# What are the mean, median, and mode of each column?
# Identify all numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove unsuitable columns
unsuitable_columns = [
    'ID', 'Kidhome', 'Complain', 'AcceptedCmp1', 'AcceptedCmp2',
    'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response',
    'Z_CostContact', 'Z_Revenue'
]

# Keep only columns suitable for mean/median/mode
suitable_columns = [col for col in numeric_columns if col not in unsuitable_columns]

# Calculate statistics
mean_values = df[suitable_columns].mean()
median_values = df[suitable_columns].median()
mode_values = df[suitable_columns].mode().iloc[0]

# Print results
print("=== Mean Values ===")
print(mean_values)
print("\n=== Median Values ===")
print(median_values)
print("\n=== Mode Values ===")
print(mode_values)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values[missing_values > 0])
```

```python
# Encode any categorical variables
# List of categorical columns to encode
categorical_cols = ['Education', 'Marital_Status']

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Preview the updated DataFrame
print(df_encoded.head())
```

```python
#Check inconsistent or unreleastic data
# Filter out customers born before 1900 or after 2005
unrealistic_ages = df[(df['Year_Birth'] < 1900) | (df['Year_Birth'] > 2005)]
print(f"Unrealistic birth years found: {unrealistic_ages.shape[0]}")
```

```python
#Check for negative values in purchase or income columns
columns_to_check = [
    'Income', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth'
]

negative_values_rows = df[(df[columns_to_check] < 0).any(axis=1)]
print(f"Rows with negative values: {negative_values_rows.shape[0]}")
```

```python
#  Check for customers with zero purchases and zero spending
product_columns = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]

purchase_columns = [
    'NumDealsPurchases', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases'
]

zero_purchases = df[
    (df[product_columns].sum(axis=1) == 0) &
    (df[purchase_columns].sum(axis=1) == 0)
]
print(f"Customers with zero spending and purchases: {zero_purchases.shape[0]}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Strongly Disagree': [5, 6, 6, 8, 16],
    'Disagree': [5, 11, 14, 32, 25],
    'Neutral': [30, 35, 45, 20, 32],
    'Agree': [27, 28, 15, 23, 18],
    'Strongly Agree': [33, 20, 20, 17, 9]
}

index = ['Survey item E', 'Survey item D', 'Survey item C', 'Survey item B', 'Survey item A']

df = pd.DataFrame(data, index=index)
df
```

```python
categories = df.columns.tolist()
colors = ['black', 'dimgray', 'lightgray', 'steelblue', 'navy']

fig, ax = plt.subplots(figsize=(10, 6))
left = [0] * len(df)

# Draw each category as a stacked horizontal bar
for i, category in enumerate(categories):
    values = df[category]
    ax.barh(df.index, values, left=left, color=colors[i], label=category)
    # Update "left" for the next bar segment
    left = [left[j] + values.iloc[j] for j in range(len(values))]

# Step 3: Formatting
ax.set_xlim(0, 100)
ax.set_xlabel('Percent of total')
ax.set_title('Survey results')
ax.invert_yaxis()  # To match visual order from top to bottom
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=5, frameon=False)
ax.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
